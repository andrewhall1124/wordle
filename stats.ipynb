```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wordle.csv", header=None, names=['word', 'occurence', 'day'])
df
```

```python
min_occurence = df[df['day'].notna()]['occurence'].min()
min_occurence
```

```python
filtered = df[df['occurence'] > min_occurence]
filtered
```

```python
filtered.to_csv("remaining_words.csv", index=False)
```
